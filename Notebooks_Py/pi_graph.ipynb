```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df2=pd.read_csv("Datos/pi_2d.csv")
df3=pd.read_csv("Datos/pi_3d.csv")

i=0
df2=df2.iloc[i:,:]
df3=df3.iloc[i:,:]

plt.title("Pi monte carlo\n 2D y 3D")
plt.errorbar(df2["n"],df2["pi"],yerr=df2["error"],marker="o",ls="",ecolor="red",label="2D")
plt.errorbar(df3["n"],df3["pi"],yerr=df3["error"],marker="o",ls="",ecolor="red",label="3D")
plt.xlabel("n")
plt.ylabel("$A_{sp}$/$A_{sq}$")
plt.legend()
plt.show()
```
